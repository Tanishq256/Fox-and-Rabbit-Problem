```python
import matplotlib.pyplot as plt
from math import sqrt
```

```python
Sf=1
Sr=1.1
dt = 0.1 #time change
tmax = 10 # seconds
Nsteps = (int)(tmax/dt) # this needs to be an integer
```

```python
Rabbitx = []
Rabbity = []
Rabbitx=[5]
Rabbity=[0]

for n in range(Nsteps):
    Rabbitx.append(5)
    Rabbity.append(Sr*n*dt)
    
#print Rabbitx


#print Rabbity

myplot = plt.plot([Rabbitx],[Rabbity],'ro')
plt.show()
```

```python
Foxx =[]
Foxy = []
Foxx=[0]
Foxy=[0]

for n in range(Nsteps):
    a = Rabbitx[n]-Foxx[n]
    b = Rabbity[n]-Foxy[n] 
    h = sqrt(a**2+b**2)
    unit_a = a/h*Sf*dt    #speed of fox is 1         Speed of Fox and Rabbit are now Sf and Sr respectively
    unit_b = b/h*Sf*dt    #Speed off fox is 1
    Foxx.append(Foxx[n]+unit_a)
    Foxy.append(Foxy[n]+unit_b)


plt.plot([Foxx],[Foxy],'bo',[Rabbitx],[Rabbity],'ro')
plt.show() 
```
